# Encryption and Decryption
### A Program to take input from user and encrypt it and vice versa
#### The encryption process involves shifting the letters of the word by the user entered key and then reversing it to produce the encrypted word. Similarly, we decrypt the word.
##### Encryption is widely used in the field of cryptography and transferring sectret messages

```python
class EncDec:
    
    def encrypt(): #to encrypt the text
        
        while 1:
            
            text= input("Enter the text to be encrypted: ")
            shift= int(input("Enter the key: "))
            encryptedtxt=""
        
            if shift>0 and shift<26:
            
                for i in range (0, len(text)): #shifting each letter and then reversing the encrypted word
                    if text[i].isupper():
                        a= chr((ord(text[i])-65+shift)%26+65)
                        encryptedtxt= a+encryptedtxt
                    elif text[i].islower():
                        a= chr((ord(text[i])-97+shift)%26+97)
                        encryptedtxt= a+encryptedtxt
                    else:
                        encryptedtxt= text[i]+encryptedtxt
        
                print("ENCRYPTED TEXT: "+encryptedtxt)
                break
            else:
                print("SHIFT CAN BE ONLY FROM 1 to 25")
    #end encrypt()
                
    def decrypt(): #to decrypt the text
        while 1:
            text= input("Enter the text to be decrypted: ")
            shift= int(input("Enter the key: "))
            decryptedtxt=""
        
            if shift>0 and shift<26:
            
                for i in range (0, len(text)): #shifting each letter and then reversing the decrypted word
                    if text[i].isupper():
                        a= chr((ord(text[i])-65-shift+26)%26+65)
                        decryptedtxt= a+decryptedtxt
                    elif text[i].islower():
                        a= chr((ord(text[i])-97-shift+26)%26+97)
                        decryptedtxt= a+decryptedtxt
                    else:
                        decryptedtxt= text[i]+decryptedtxt
        
                print("DECRYPTED TEXT: "+decryptedtxt)
                break
            else:
                print("SHIFT CAN BE ONLY FROM 1 to 25")
    #end decrypt()    
    if __name__=="__main__":
        
        while 1:
            
            choice= int(input("Enter your choice\n1.Encrypt\n2.Decrypt\n3.Exit\n"))
       
            if choice==1:
                encrypt()
                break
            elif choice==2:
                decrypt()
                break
            elif choice==3:
                exit()
                break
            else: print("INVALID CHOICE. PLEASE RE-ENTER YOUR CHOICE")
    
   #end main()
        
        
```
